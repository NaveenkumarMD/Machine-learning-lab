```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/NaveenkumarMD/Machine-learning-lab/main/Bayes%20theorem/data.csv")
df
```

```python
give_birth_arr=df["Give Birth"].to_numpy()
can_fly_arr=df["Can Fly"].to_numpy()
live_in_water_arr=df["Live in Water"].to_numpy()
live_in_water_arr=np.array([i.strip() for i in live_in_water_arr])
have_legs_arr=df["Have Legs"].to_numpy()
class_arr=df["Class"].to_numpy()
live_in_water_arr
```

```python
gb_unique=np.unique(give_birth_arr)
cf_unique=np.unique(can_fly_arr)
lw_unique=np.unique(live_in_water_arr)
hl_unique=np.unique(have_legs_arr)
cl_unique=np.unique(class_arr)
print(gb_unique)
print(cf_unique)
print(lw_unique)
print(hl_unique)
print(cl_unique)
totallen=len(give_birth_arr)
totallen
```

```python
from collections import Counter
gb_count=Counter(give_birth_arr)
cf_count=Counter(can_fly_arr)
lw_count=Counter(live_in_water_arr)
hl_count=Counter(have_legs_arr)
cl_count=Counter(class_arr)
print(gb_count)
print(cf_count)
print(lw_count)
print(hl_count)
print(cl_count)
```

```python
gb_prob={
    i:{} for i in gb_unique
}
cf_prob={
    i:{} for i in cf_unique
}
lw_prob={
    i.strip():{} for i in lw_unique
}
hl_prob={
    i:{} for i in hl_unique
}
for i in range(totallen):
  if class_arr[i] not in gb_prob[give_birth_arr[i]]:
    gb_prob[give_birth_arr[i]][class_arr[i]]=1/cl_count[class_arr[i]]
  else:
    gb_prob[give_birth_arr[i]][class_arr[i]]+=1/cl_count[class_arr[i]]
for i in range(totallen):
  if class_arr[i] not in cf_prob[can_fly_arr[i]]:
    cf_prob[can_fly_arr[i]][class_arr[i]]=1/cl_count[class_arr[i]]
  else:
    cf_prob[can_fly_arr[i]][class_arr[i]]+=1/cl_count[class_arr[i]]

for i in range(totallen):
  if class_arr[i] not in lw_prob[live_in_water_arr[i]]:
    lw_prob[live_in_water_arr[i]][class_arr[i]]=1/cl_count[class_arr[i]]
  else:
    lw_prob[live_in_water_arr[i]][class_arr[i]]+=1/cl_count[class_arr[i]]
for i in range(totallen):
  if class_arr[i] not in hl_prob[have_legs_arr[i]]:
    hl_prob[have_legs_arr[i]][class_arr[i]]=1/cl_count[class_arr[i]]
  else:
    hl_prob[have_legs_arr[i]][class_arr[i]]+=1/cl_count[class_arr[i]]
print(gb_prob)
print(hl_prob)
print(cf_prob)
print(lw_prob)
```

```python
#testing
a='yes'
b="no"
c="yes"
d="no"
mam_prob=gb_prob[a]['mammals']+hl_prob[b]['mammals']+lw_prob[c]["mammals"]+cf_prob[d]["mammals"]
non_mam_prob=gb_prob[a]['non-mammals']+hl_prob[b]['non-mammals']+lw_prob[c]["non-mammals"]+cf_prob[d]["non-mammals"]
if non_mam_prob>mam_prob:
  print("non-mammals")
else:
  print("mammals")
```
